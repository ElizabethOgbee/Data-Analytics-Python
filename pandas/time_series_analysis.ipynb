STEP 2- RESAMPLING DATA

THIS IS USED TO CHANGE THE FREQUENCY OF YOUR TIME SERIES- 

TWO TYPES OF RESAMPLING INCLUDES:
- DOWNSAMPLING = GOING FROM HIGHER FREQUENCY TO LOWER FREQUENCY (eg. daily -> monthly)
- UPSAMPLING = GOING FROM LOWER FREQUENCY TO HIGHER FREQUENCY (eg., monthly).

RESAMPLING LET'S YOU SUMMARIZE TIME-BASED DATA( USING MEAN, SUM, COUNT, ETC)
