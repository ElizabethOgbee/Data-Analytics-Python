Today, I'll be learning about FILTERING, GROUPING AND HOW TO MERGE BOTH WITH CONDITIONAL STATEMENT

```python
# 1. LOAD FILE#
import pandas as pd

df = pd.read_csv("supmkt.csv")
df.head()
```

```python
#2. FILTERING WITH .LOC[]
# - GOAL:
# IS TO SELECT THE ROWS OF A CERTAIN CONDITION
# -- FILTER WHERE SUPERMARKET SIZE IS 'MEDIUM'#

df_size = df.loc[df['Supermarket_Size'] == 'Medium']
df_size.head()

```

```python
# 2B: FILTER WHERE Product_Supermarket_Sales IS GREATER THAN 5000#
df_product_supermarket_amount = df.loc[df['Product_Supermarket_Sales'] > 5000]
df_product_supermarket_amount.head()
```

```python
# 2C. COMBINING BOTH df_size AND product_supermarket_amount < 5000 #
df_filtered = df.loc[(df['Supermarket_Size'] == 'Medium') & (df['Product_Supermarket_Sales'] > 5000)]
df_filtered.head()
```
