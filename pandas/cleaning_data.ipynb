CLEANING DATA

THIS STAGE IS ALL ABOUT MAKING MY DATA NEAT AND READY FOR ANALYSIS. MESSY DATA CAN LEADTO WRONG CONCLUSIONS. 

THE STEPS:
1- REMOVING NULLS(MISSING DATA)

```python
#CHECK FOR NULLS FIRST
import pandas as pd
import re

df = pd.read_csv('supmkt.csv')

df.isnull().sum()
```

```python
#REMOVE ROWS WITH ANY MISSING VALUE:
df.dropna()
```

```python
#REMOVE COLUMNS WITH MISSING VALUES
df.dropna(axis=1)
df.dropna()
```

```python
#FILL MISSING VALUES WITH A SPECIFIC VALUE(EG. 0, 'UNKNOWN', OR A COLUMN MEAN)
df.fillna(0)
df.fillna('unkwown')
df['Product_Supermarket_Sales'].fillna(df['Product_Supermarket_Sales'].mean())
```

```python
| Code                                 | Best For              | What It Does                                         |
| ------------------------------------ | --------------------- | ---------------------------------------------------- |
| `df.fillna(0)`                       | Numeric data          | Replaces all NaNs with 0                             |
| `df.fillna('unknown')`               | Text/categorical data | Replaces all NaNs with 'unknown'                     |
| `df['col'].fillna(df['col'].mean())` | Numeric data          | Fills NaNs in one column using that column's average |
```

TRANSFORMING DATA

Today, we will learn how to transform our clean data in various ways possible. The effect of transforming data is that it will help us reshap our data to make it easier to understand or analyse

```python
# 1. CREATING NEW COLUMN
df['Discounted_Price'] = df['New_Product_Price'] * 0.8
# This creates a new column where every product price is reduced by 20%.
```

```python
#2. Use a function on a column .apply()
# 
# YOU CAN USE .APPLY TO RUN A FUNCTION ON EVERY VALUE IN A COLUMN.
# 
# EG: INCREASE PRICE BY 10%#

df['New_Price'] = df['Product_Price'].apply(lambda x: x * 1.10)
```

```python
#EXAMPLE 2: Label expensive product:
df['Price_Label']= df['product_price'].apply(lambda x: 'High' if x > 100 else 'Low')
```

```python
#BINDING OR CATEGORISING DATA
# Binding means turning numbers into categories or labels.
# Usenpd.cut() for fixed ranges and pd,qcut() for equal sized groups
pd.cut() #-FIXED BINS
eg:
bins = [0, 50, 100, 150]
labels = ['Low', 'Medium', 'High']
df['price_Group'] = pd.cut(df['product_price'], bins=bins, labels=labels)

#This breaks product prices into 3 groups.#
```

```python
#PD.QCUT()- EQUAL SIZEED GROUPS
df['Price_Quartile'] = pd.qcut(df['product_price'], q=4)
#THIS SPLIT PRODUCT INTO 4 EQUAL GROUPS(QUARTILES).
```

```python
#4. .map() - REPLACE sPECIFIC vALUES
# 
# USE THIS WHEN YOU WANT TO RENAME OR RELABEL VALUES IN A COLUMN
# EX:#
df['Fat_Content'] = df['Fat_Content'].map({
    'LF': 'Low Fat',
    'low fat': 'Low Fat',
    'reg': 'Regular'
})
```

```python
#5. .REPLACE() - MORE FLECIBLE REPLACEMENT
# .replace() works like .map() but it's more flexible.
# 
# eg:#
df['supermarket_size'].replace(['Small', 'Medium', 'High'])
```

✅ Summary
| Tool         | What It Does                           | Example Use                              |
| ------------ | -------------------------------------- | ---------------------------------------- |
| New Columns  | Make new data from existing ones       | Discounts, tax, combined fields          |
| `.apply()`   | Apply a function to each value         | Raise price, classify as cheap/expensive |
| `pd.cut()`   | Divide data into fixed range groups    | Group prices into Low/Med/High           |
| `pd.qcut()`  | Divide data into equal-sized groups    | Create quartiles                         |
| `.map()`     | Replace specific values                | Rename values like ‘LF’ to ‘Low Fat’     |
| `.replace()` | Replace one or many values in a column | Change names, labels, symbols            |
